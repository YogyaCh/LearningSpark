```python
from pyspark.sql import SparkSession
```

The **SparkSession** class is the entry point to programming with Apache Spark. It allows you to create DataFrames and execute SQL queries, and it provides access to all of Spark's functionalities in a unified and straightforward way.

```python
spark = SparkSession.builder.appName("Python Spark SQL basic example").getOrCreate()
```

Reading json data as a dataframe

```python
df = spark.read.json("./data/people.json")
df.show()
```

```python
df.printSchema()
```

```python
df.select('name', df['age'] + 1).show()
```

```python
df.filter(df['age'] > 21).show()
```

```python
df.groupBy("age").count().show()
```

This provides list of APIs provided for data frames: https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql.html#dataframe-apis

```python
df.createOrReplaceTempView("people")

sqlDF = spark.sql("SELECT * FROM people")
sqlDF.show()
```
